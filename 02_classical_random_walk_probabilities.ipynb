```python
import numpy as np
import matplotlib.pyplot as plt

def find_b_probability(num_vertices, num_steps):
    # Transition matrix for complete symmetric graph
    P = np.full((num_vertices, num_vertices), 1 / (num_vertices - 1))
    np.fill_diagonal(P, 0)
    P[1, :] = 0  # Once at vertex b, you cannot move out
    P[1, 1] = 1

    # Initial state: equal probability of starting at any vertex
    state = np.full(num_vertices, 1 / num_vertices)

    probabilities_b = [state[1]]  # Store the initial probability of being at vertex b

    for _ in range(num_steps - 1):
        state = np.dot(state, P)
        probabilities_b.append(state[1])

    return probabilities_b
```

```python
num_vertices = 4
num_steps = 40
probabilities = find_b_probability(num_vertices, num_steps)

# Plot the probabilities
plt.plot(range(num_steps), probabilities, marker='o',color='red')
plt.xlabel('Number of Steps')
plt.ylabel('Probability of Being at a Fixed Vertex')
plt.title(f'Random Walk on Complete Symmetric Graph G=({num_vertices}, {num_vertices * (num_vertices - 1) // 2})')
#plt.grid()
plt.savefig('quantum_random_walk_2.pdf')
plt.show()
```

```python
print("Probabilities of being at vertex b after each step:")
for step, prob in enumerate(probabilities):
    print(f"Step {step}: {prob:.4f}")
```
